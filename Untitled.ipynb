```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Step 2: Load the dataset
df = pd.read_csv("Uber-Jan-Feb-FOIL.csv")
```

```python
# Step 3: Explore the data
print(df.head())
print(df.info())
print(df.describe())

```

```python
# Step 4: Convert Date/Time column to datetime
df['date'] = pd.to_datetime(df['date'],errors='coerce')
```

```python
# Add new time-related columns
df['day'] = df['date'].dt.day
df['weekday'] = df['date'].dt.day_name()
df['week'] = df['date'].dt.isocalendar().week
df['month'] = df['date'].dt.month_name()
```

```python
# Total trips per day
plt.figure(figsize=(12, 5))
df.groupby('date')['trips'].sum().plot()
plt.title('Daily Uber Trips')
plt.xlabel('Date')
plt.ylabel('Number of Trips')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# Total trips per base
plt.figure(figsize=(8, 4))
df.groupby('dispatching_base_number')['trips'].sum().sort_values(ascending=False).plot(kind='bar', color='orange')
plt.title('Total Trips per Base')
plt.xlabel('Base')
plt.ylabel('Trips')
plt.tight_layout()
plt.show()
```

```python
# Active vehicles vs trips scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='active_vehicles', y='trips', hue='dispatching_base_number')
plt.title('Trips vs Active Vehicles')
plt.xlabel('Active Vehicles')
plt.ylabel('Trips')
plt.tight_layout()
plt.show()
```

```python
# Average trips by weekday
plt.figure(figsize=(10, 5))
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df.groupby('weekday')['trips'].mean().reindex(weekday_order).plot(kind='bar', color='green')
plt.title('Average Trips by Weekday')
plt.ylabel('Average Trips')
plt.tight_layout()
plt.show()
```

```python
# Total trips per base
plt.figure(figsize=(8, 5))
df.groupby('dispatching_base_number')['trips'].sum().sort_values().plot(kind='barh', color='purple')
plt.title("Total Uber Trips by Dispatching Base")
plt.xlabel("Total Trips")
plt.ylabel("Base")
plt.tight_layout()
plt.show()
```

```python
#Compare trip trends over time by base.
pivot = df.pivot_table(index='date', columns='dispatching_base_number', values='trips', aggfunc='sum')

pivot.plot(figsize=(12, 5))
plt.title("Trips Over Time by Base (Jan-Feb 2015)")
plt.xlabel("Date")
plt.ylabel("Number of Trips")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#Trip intensity by day of the week and base.
df['weekday'] = df['date'].dt.day_name()

# Pivot table
heatmap_data = df.pivot_table(index='weekday', columns='dispatching_base_number', values='trips', aggfunc='mean')

# Reorder weekdays
ordered = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
heatmap_data = heatmap_data.reindex(ordered)

sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Average Trips per Base by Weekday")
plt.tight_layout()
plt.show()
```

```python
#Trips vs Active Vehicles with bubble size for visual impact.
# Group data by base
grouped = df.groupby('dispatching_base_number').agg({'trips': 'sum', 'active_vehicles': 'sum'}).reset_index()

plt.figure(figsize=(8, 6))
plt.scatter(grouped['active_vehicles'], grouped['trips'], s=grouped['trips'] / 10, alpha=0.6)

for i in range(len(grouped)):
    plt.text(grouped['active_vehicles'][i], grouped['trips'][i], grouped['dispatching_base_number'][i])

plt.xlabel("Total Active Vehicles")
plt.ylabel("Total Trips")
plt.title("Trips vs Active Vehicles per Base (Bubble Size = Trips)")
plt.tight_layout()
plt.show()
```

```python
# Trips per Week (Line Plot)
df['week'] = df['date'].dt.isocalendar().week

weekly_trips = df.groupby('week')['trips'].sum()

weekly_trips.plot(marker='o', color='teal', figsize=(8, 4))
plt.title("Total Weekly Trips (JanFeb 2015)")
plt.xlabel("Week Number")
plt.ylabel("Total Trips")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap (Trips vs Vehicles)
sns.heatmap(df[['trips', 'active_vehicles']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()

```
